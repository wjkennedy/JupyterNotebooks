# Customer Journey DSL Builder

This notebook helps capture how a customer describes the journey for their request or issue and exports it as JSON that can be reused for diagramming or workflow generation in systems like Jira or Jira Service Management.

```
import json
import ipywidgets as widgets
from IPython.display import display
```

## Configure journey terms

Provide the term the customer uses for the item being tracked, give the overall journey name, and specify how many stages exist. Click **Create Stage Inputs** to describe each stage.

```
artifact_name = widgets.Text(description='Item term', placeholder='e.g., Case')
journey_name = widgets.Text(description='Journey name', placeholder='e.g., Support Request')
num_stages = widgets.IntText(description='Number of stages', value=3)
create_btn = widgets.Button(description='Create Stage Inputs')
setup_ui = widgets.VBox([artifact_name, journey_name, num_stages, create_btn])
display(setup_ui)

stage_widgets = []
stages_box = widgets.VBox()
output = widgets.Output()
display(stages_box, output)


def on_create_clicked(b):
    stage_widgets.clear()
    children = []
    for i in range(num_stages.value):
        stage_term = widgets.Text(description=f'Stage {i+1} term')
        customer_action = widgets.Text(description='Customer action')
        internal_process = widgets.Text(description='Internal process')
        stage_widgets.append({'stage_term': stage_term, 'customer_action': customer_action, 'internal_process': internal_process})
        children.extend([stage_term, customer_action, internal_process])
    generate_btn = widgets.Button(description='Generate JSON')
    def on_generate_clicked(btn):
        data = {
            'artifact': artifact_name.value,
            'journey_name': journey_name.value,
            'stages': [{
                'stage_term': w['stage_term'].value,
                'customer_action': w['customer_action'].value,
                'internal_process': w['internal_process'].value
            } for w in stage_widgets]
        }
        output.clear_output()
        with output:
            print(json.dumps(data, indent=2))
    generate_btn.on_click(on_generate_clicked)
    children.append(generate_btn)
    stages_box.children = children

create_btn.on_click(on_create_clicked)
```

After filling in all fields, click **Generate JSON** to view the DSL representation of the customer journey.
